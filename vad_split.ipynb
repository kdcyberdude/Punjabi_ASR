```python
import torch
from speech_utils import load_audio, SAMPLE_RATE
import sys
sys.path.append('..')
from utils import vad as vad
import IPython.display as ipd

chunk_size = 15
device = 'cpu'

vad_params = {
    "vad_onset": 0.500,
    "vad_offset": 0.363
}
vad_model = vad.load_vad_model(torch.device(device), use_auth_token=None, **vad_params)

def vad_segments(audio: str):
    audio = load_audio(audio)

    vad_segments = vad_model({"waveform": torch.from_numpy(audio).unsqueeze(0), "sample_rate": SAMPLE_RATE})
    vad_segments = vad.merge_chunks(
        vad_segments,
        chunk_size,
        onset=vad_params["vad_onset"],
        offset=vad_params["vad_offset"],
    )

    return audio, vad_segments


audio, segments = vad_segments(audio='example.wav')
print('\n\n')
print(segments)
```

```python

audio, segments = vad_segments(audio='example.wav')
for s in segments:
    ipd.display(ipd.Audio(audio[int(s['start']*SAMPLE_RATE):int(s['end']*SAMPLE_RATE)], rate=SAMPLE_RATE))
    print(s['end'] - s['start'])
    print('-------------------------------------')
```
